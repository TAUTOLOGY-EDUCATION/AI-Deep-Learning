```python
import pickle
from glob import glob

import numpy as np
import pandas as pd
from PIL import Image
```

```python
saved_model = pickle.load(open("sign_language_model.pickle", "rb"))
```

```python
clf = saved_model[0]
scaler = saved_model[1]
```

```python
img_path = glob("dataset/B/*")[0]
incoming_data = Image.open(img_path)

incoming_data
```

```python
incoming_data = incoming_data.resize([64, 64])
incoming_data = np.array(incoming_data).reshape(1, -1)
incoming_data = scaler.transform(incoming_data)

incoming_data
```

# Prediction

```python
predicted_target = clf.predict(incoming_data)
```
